```python
def create_dummies(list_of_columns, dateframe):
    new_dfs = []
    for x in list_of_columns:
        dummies = pd.get_dummies(dateframe[x])
        new_dfs.append(dummies)
    return(new_dfs)
```

```python
str_columns = df8.select_dtypes(include=['object']).columns
```

```python
dumms = create_dummies(str_columns, df8)
```
